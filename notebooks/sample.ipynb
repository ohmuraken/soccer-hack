```python
premier15 =pd.read_csv("../data/england/premier_2015_2016.csv")
# 欠損地が無いかをチェック
premier15.isnull().values.any()
```

```python
teams = np.unique(premier15["HomeTeam"])
target_team = "Liverpool"
target_team_match = premier15[(premier15["HomeTeam"]==target_team) | (premier15["AwayTeam"]==target_team)]
target_team_home_match = target_team_match[target_team_match["HomeTeam"]==target_team]
target_team_away_match = target_team_match[target_team_match["AwayTeam"]==target_team]
all_match_num = target_team_match.shape[0]
```

```python
def get_premier15_data(target_team):
    target_team_match = premier15[(premier15["HomeTeam"]==target_team) | (premier15["AwayTeam"]==target_team)]
    target_team_home_match = target_team_match[target_team_match["HomeTeam"]==target_team]
    target_team_away_match = target_team_match[target_team_match["AwayTeam"]==target_team]
    all_match_num = target_team_match.shape[0]

    # Homeでの戦績
    home_win_count = 0
    home_lose_count = 0
    home_draw_count = 0
    for result in target_team_home_match["FTR"]:
        if result == "H":
            home_win_count += 1
        elif result == "A":
            home_lose_count += 1
        elif result == "D":
            home_draw_count += 1
    home_win_rate = home_win_count / all_match_num
#     print("home_win_count: {}".format(home_win_count))
#     print("home_lose_count: {}".format(home_lose_count))
#     print("home_draw_count: {}".format(home_draw_count))
#     print("---------")
    
    # Awayでの勝率
    away_win_count = 0
    away_lose_count = 0
    away_draw_count = 0
    for result in target_team_away_match["FTR"]:
        if result == "A":
            away_win_count += 1
        elif result == "H":
            away_lose_count += 1
        elif result == "D":
            away_draw_count += 1
    away_win_rate = away_win_count / all_match_num
#     print("away_win_count: {}".format(away_win_count))
#     print("away_lose_count: {}".format(away_lose_count))
#     print("away_draw_count: {}".format(away_draw_count))
#     print("---------")
    
    print("total win: {}".format(home_win_count+away_win_count))
    print("total draw: {}".format(home_draw_count+away_draw_count))
    print("total lose: {}".format(home_lose_count+away_lose_count))
```

```python
target_teams = teams
for team in target_teams:
    print("================")
    print("team: {}".format(team))
    get_premier15_data(team)
```
